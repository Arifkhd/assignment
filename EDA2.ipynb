Use Label Encoding for categorical variables with more than 5 categories.

```python
def label_encoder(df,columns):
    df = df.copy()
    encoders = {}
    
    for column in columns:
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        encoders[column] = le
    return df, encoders
```

```python
columns_to_encode = ['workclass', 'education','marital_status','occupation','relationship','native_country']
```

```python
df, encoders = label_encoder(df,columns_to_encode)
```

3. Feature Engineering:

a. Interaction Features:

Interaction features are created by combining two or more existing features. They can capture relationships between features that are not apparent when using the features individually.

Rationale: Higher education often correlates with better job positions, which might also influence the number of hours worked per week. By creating an interaction feature, we can capture this relationship and its potential effect on income.

```python
#Combining 'education-num' and 'hours-per-week'
df['education'] = df['education_num'] * df['hours_per_week']
```

```python
#'capital-gain' to 'capital-loss' ratio
df['capital_gain_loss_ratio'] = df['capital_gain'] / (df['capital_loss'] + 1)  # Adding 1 to avoid division by zero
```

Applying Transformations

a. Log Transformation:

Log transformation is often used to reduce skewness and make the data distribution more normal. This transformation is particularly useful for features with long tails or large variances.

Rationale: The 'capital-gain' feature may have a right-skewed distribution due to a small number of individuals reporting very high capital gains. Applying a log transformation can normalize the distribution, making it more symmetric and reducing the influence of extreme values on the model.

```python
#capital_gain
df['log_capital_gain'] = np.log(df['capital_gain'] + 1)  # Adding 1 to handle zero values
```
